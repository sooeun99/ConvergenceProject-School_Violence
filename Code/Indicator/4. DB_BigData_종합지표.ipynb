# 환경지표 + 위해지표 + 경감지표 +학교정보 

```python
data1 = pd.merge(school_data, Environmental_indicators, on=['시군구'], how='outer', sort=True)
data2 = pd.merge(data1, Hazard_indicators, on=['학교명'], how='outer', sort=True)
data3 = pd.merge(data2, Alleviate_indicators, on=['학교명'], how='outer', sort=True)
data3= data3[['학교명', '시군구', '환경지표_score', '환경지표_cluster', '환경지표_학교폭력위험도','위해지표_score',
       '위해지표_cluster','위해지표_학교폭력위험도', '경감지표_score','경감지표_cluster','경감지표_학교폭력위험도']]
data3
```

```python
data3['total_score'] = data3['환경지표_score']*0.457842 + data3['위해지표_score']*0.263726 + data3['경감지표_score']*0.278432
data3
```

```python
percent = [0.20, 0.40, 0.60, 0.80]
```

```python
data3['total_score'].describe(percentiles = percent) 
```

```python
def group(score):
    if score>44.139709: return '매우위험'
    elif score>35.089207: return '위험'
    elif score>26.401918: return '보통'
    elif score>16.572482: return '안전'
    else: return '매우안전'

score_group = data3['total_score'].apply(lambda x:group(x))
data3['종합_위험도'] = score_group
data3
# total_Score 기준 상위20%
```

```python
data3 = data3[['학교명','시군구','환경지표_score','위해지표_score','경감지표_score','total_score','종합_위험도']]
data3['종합_위험도'].value_counts()
```
